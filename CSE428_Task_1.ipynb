```python
import numpy as np
from skimage import io, color, exposure, img_as_float
import matplotlib.pyplot as plt
```

#Problem 1

```python
# Load the "cat.png" image
I = io.imread("/content/cat.png")

# Normalize the image
I = img_as_float(I)

# Calculate the height and the width of the image

height = I.shape[0]
width = I.shape[1]

print("Height of the image - ", height)
print("Width of the image - ", width)
```

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 5))

# RGB image
axes[0].imshow(I)
axes[0].axis("off")
axes[0].set_title("RGB image")

# Red channel
axes[1].imshow(I[:, :, 0], cmap="gray")
axes[1].axis("off")
axes[1].set_title("Red")

# Green channel
axes[2].imshow(I[:, :, 1], cmap="gray")
axes[2].axis("off")
axes[2].set_title("Green")

# Blue channel
axes[3].imshow(I[:, :, 2], cmap="gray")
axes[3].axis("off")
axes[3].set_title("Blue")

# Alpha channel
axes[4].imshow(I[:, :, 3], cmap="gray")
axes[4].axis("off")
axes[4].set_title("Alpha");
```

```python
# Flip the image along the x-axis
I_flipped = I[::-1, :, :]

plt.imshow(I_flipped)
plt.axis('off');
```

#Problem 3

Run the next cell to convert the image to grayscale.

```python
I = I[:,:,:3] # alpha channel has to be dropped
I = color.rgb2gray(I)
plt.imshow(I, cmap = 'gray')
plt.axis('off');
```

Now calculate the PDF for the grayscale image

```python
# Calculate the PDF for the image
hist, bins_hist = exposure.histogram(I, nbins=256, normalize = True)

plt.plot(bins_hist, hist);
plt.xlabel("Pixel values")
plt.title('Probability Density Function')
plt.ylabel("Percentage");
```
